```python
corpus_fname = '2016-10-20_article_all_normed.txt' # YOUR CORPUS

import sys
sys.path.append('/mnt/lovit/git/soynlp/')
corpus_fname = '/mnt/lovit/works/fastcampus_text_ml/1st/data/corpus_10days/news/2016-10-20_article_all_normed.txt'

from soynlp.utils import DoublespaceLineCorpus
corpus = DoublespaceLineCorpus(corpus_fname, iter_sent=False, num_doc=1000)

for n_doc, doc in enumerate(corpus):
    print('[doc={}]: {}\n'.format(n_doc, doc[:200].strip()))
    if n_doc >= 10:
        break
```

## Training model

```python
from wordpiecemodel import BytePairEncoder

encoder = BytePairEncoder(1000)
encoder.train(corpus)
```

```python
for n_doc, doc in enumerate(corpus):
    if not doc:
        continue
        
    print('[doc={}]: {}'.format(n_doc, doc[:200].strip()))
    print('[BPE tokens]: {} ... \n'.format(''.join(encoder.tokenize(doc)[:300])))
    
    if n_doc >= 10:
        break
```

## I/O: Save and load model

```python
encoder.save('bpe1000')
loaded_encoder = BytePairEncoder()
loaded_encoder.load('bpe1000')
```

```python
for n_doc, doc in enumerate(corpus):
    if not doc:
        continue
        
    print('[doc={}]: {}'.format(n_doc, doc[:200].strip()))
    print('[BPE tokens]: {} ... \n'.format(''.join(loaded_encoder.tokenize(doc)[:300])))
    
    if n_doc >= 5:
        break
```

## Model with large number of units

```python
encoder_large = BytePairEncoder(5000)
encoder_large.train(corpus)
```

```python
for n_doc, doc in enumerate(corpus):
    if not doc:
        continue
        
    print('[doc={}]: {}'.format(n_doc, doc[:200].strip()))
    print('[BPE 1000 tokens]: {} ... '.format(''.join(encoder.tokenize(doc)[:300])))
    print('[BPE 5000 tokens]: {} ... \n'.format(''.join(encoder_large.tokenize(doc)[:300])))
    
    if n_doc >= 10:
        break
```
